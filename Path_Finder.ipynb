```python
import matplotlib.pyplot as plt
import numpy as np
import random
from scipy.spatial import distance

x=[]
y=[]
x1=[]
y1=[]

'''x=[5,10,7,3,6]
y=[1,3,2,10,6]
x1=[4,9]
y1=[2,6]'''

for i in range(0,50):
    x.append(random.randint(1,50))
    y.append(random.randint(1,50))
    
for i in range(0,7):
    x1.append(random.randint(1,48))
    y1.append(random.randint(1,48))

nfp_x=[]
nfp_y=[]

min = distance.euclidean((0,0), (x[0],y[0]) )
for i in range(0,50):
    if (distance.euclidean((0,0), (x[i],y[i]) ) < min):
        min = distance.euclidean((0,0), (x[i],y[i]))
        aag_x_1=x[i]
        aag_y_1=y[i]

nfp_x.append(aag_x_1)
nfp_y.append(aag_y_1)

for j in range(0,7):
    min = distance.euclidean((x1[j],y1[j]), (x[0],y[0]) )
    for i in range(0,50):
        if (distance.euclidean((x1[j],y1[j]), (x[i],y[i]) ) < min):
            min = distance.euclidean((x1[j],y1[j]), (x[i],y[i]))
            aag_x_1=x[i]
            aag_y_1=y[i]

    nfp_x.append(aag_x_1)
    nfp_y.append(aag_y_1)

plt.plot(nfp_x,nfp_y, marker='o', color='y', label="Path")
plt.plot(x,y, 'o', color='g', label="Working nodes" )
plt.scatter(x1,y1,s=50,alpha=0.9, marker='o', color='r',label="Down nodes")
plt.xlabel("Distance in KMs")
plt.ylabel("Distance in KMs")
#plt.legend("upper right")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```
